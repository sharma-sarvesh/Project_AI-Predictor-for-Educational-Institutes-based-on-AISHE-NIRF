```python
import pandas as pd

# Load the dataset
data = pd.read_csv("EngineeringRanking_Final.csv")

# Assuming 'Institute Id' is the unique identifier for each institute
institute_ids = data['Institute Id'].unique()

# Initialize a dictionary to store predictions
predictions = {}

# Loop through each institute
for id in institute_ids:
    # Extract data for the current institute
    institute_data = data[data['Institute Id'] == id]

    # Filter out non-numeric values from rank columns
    rank_columns = ['Rank_21', 'Rank_22', 'Rank_23']
    numeric_rank_data = institute_data[rank_columns].apply(pd.to_numeric, errors='coerce')

    # Calculate average rank change from 2016 to 2023
#     rank_change = numeric_rank_data.mean(axis=1) - numeric_rank_data['Rank_16']
    rank_change = numeric_rank_data.mean(axis=1)

    # Predict the rank for 2024 based on average change
    last_rank_2023 = numeric_rank_data['Rank_23'].iloc[-1]
    predicted_rank_2024 = last_rank_2023 + rank_change.iloc[-1]

    # Store the prediction for the current institute
    predictions[id] = predicted_rank_2024

# Create a DataFrame from the predictions dictionary
df = pd.DataFrame(predictions.items(), columns=['Institute Id', 'Predicted Rank'])

# Save the DataFrame to an Excel file
df.to_excel("predicted_ranks.xlsx", index=False)
```
